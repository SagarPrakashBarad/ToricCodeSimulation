```python
import qiskit
from qiskit_aer import AerSimulator
from matplotlib import pyplot as plt
from qiskit.compiler import transpile
```

```python
# 3-qubit GHZ state
circ = qiskit.QuantumCircuit(3)
circ.h(0)
circ.cx(0, 1)
circ.cx(1, 2)
circ.measure_all()
```

```python
circ.draw('mpl')
```

```python
aersim = AerSimulator()
```

```python
from qiskit.compiler import transpile, execute
```

```python
result_ideal = transpile(circ, aersim).result()
counts_ideal = result_ideal.get_counts(0)
```
